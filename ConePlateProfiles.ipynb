```python
#Importing libraries
import matplotlib as mpl

%config InlineBackend.figure_formats = ['svg']

mpl.rcParams.update(mpl.rcParamsDefault)
mpl.rcParams.update({'mathtext.fontset': 'cm'})
mpl.rcParams.update({'axes.labelsize': 22})
mpl.rcParams.update({'axes.titlesize': 16})
mpl.rcParams.update({'axes.linewidth': 0.5})
mpl.rcParams.update({'xtick.labelsize': 10})
mpl.rcParams.update({'ytick.labelsize': 10})

import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np #Needed for a bunch of mathematical operations.
```

```python
def vphi_1(r,theta):
    T1 = -r*Omega*(theta-np.pi/2)/theta0 
    T2 = v_slip
    T3 = 2*v_slip*(theta-np.pi/2)/theta0
    return (T2-T3+T1)

def vphi_2(r,theta):
    F = np.sin((np.pi/2)-theta0)
    T1 = Omega*r*F*np.sin(theta)/(F+1)
    T2 = v_slip
    T3 = 2*v_slip*np.sin(theta)/(F+1)
    print(np.pi/2 - theta)
    ans =(T2-T3+T1)
    print(ans)
    return ans

def vphi_3(r,theta):
    T1 = np.cos(theta0)*v_slip
    T2 = (r*Omega - 2*v_slip)*np.sin(theta)
    T3 = v_slip-r*Omega
    Denom = np.cos(theta0)-1
    ans = (T1+T2+T3)/Denom
    return ans

def Vnoslip(r,theta):
    T = r*Omega*(np.sin(theta)-1)
    N = np.cos(theta0) - 1
    ans = T/N
    return ans
```

```python
import time
tic = time.perf_counter()
theta0 = 0.7
Omega = 2
v_slip = 0.25
theta = np.linspace((np.pi/2),np.pi/2-theta0,100)
r = 1
N = 100
Vphi = np.zeros(N)
Vnslip = np.zeros(N)

for j in range(0,N):
    Vphi[j] = vphi_3(r,theta[j])
    Vnslip[j] = Vnoslip(r,theta[j])

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot( Vphi, -np.sin(theta)+1,  color= 'red', label= 'Slip')
ax.plot( Vnslip, -np.sin(theta)+1,  color= 'blue', label= 'NoSlip')
ax.set_xlabel('Velocity')
ax.set_ylabel('Height')

#ax.axis([0.01, 10000, 1000, 1e5])
ax.set_title('KBKZ Prediction')
ax.legend(loc='best',title='Legend')
ax.grid()  
toc = time.perf_counter()
print(f'Time to solution = {toc - tic:.8f} seconds')
```
